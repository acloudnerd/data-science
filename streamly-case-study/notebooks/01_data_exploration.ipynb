<h1>Value Distributions / Data Quality</h1>

```python
print("Profiles age_band value counts:")
print(profiles["age_band"].value_counts())

print("\nProfiles kids_profile value counts:")
print(profiles["kids_profile"].value_counts())

print("\nProfiles preffered_language value counts:")
print(profiles["preferred_language"].value_counts())

print("/nTitles age_rating value counts:")
print(titles["age_rating"].value_counts())

print("\nTitles category value counts (top 20):")
print(titles["category"].value_counts().head(20))

print("\nMissing values per column (profiles):")
print(profiles.isna().sum())

print("\nMissing values per column (titles):")
print(titles.isna().sum())
```
